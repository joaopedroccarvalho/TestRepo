### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
print("NГєmero de duplicatas:", df.duplicated().sum())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
df = df.drop_duplicates()
print("Linhas apГіs remoГ§ГЈo:", len(df))
print("Duplicatas restantes:", df.duplicated().sum())
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
missing = df.isnull().sum()
missing_percent = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Ausentes': missing, 'Percentual': missing_percent})
print(missing_df[missing_df['Ausentes'] > 0].sort_values('Percentual', ascending=False))
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
print("Valores ausentes em RemoteWork:", df['RemoteWork'].isnull().sum())
print("Percentual:", round(df['RemoteWork'].isnull().sum() / len(df) * 100, 2), "%")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
print(df['RemoteWork'].value_counts())
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
valor_frequente = df['RemoteWork'].mode()[0]
print("Valor mais frequente:", valor_frequente)
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
df['RemoteWork'] = df['RemoteWork'].fillna(valor_frequente)
print("Valores ausentes apГіs imputaГ§ГЈo:", df['RemoteWork'].isnull().sum())
print(df['RemoteWork'].value_counts())
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
comp_cols = ['CompTotal', 'ConvertedCompYearly']
print(df[comp_cols].describe())
print("\nValores ausentes:")
print(df[comp_cols].isnull().sum())
```
