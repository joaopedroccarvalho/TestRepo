#### Step 2: Load and Preview the Dataset
Load the dataset from the provided URL. Use df.head() to display the first few rows to get an overview of the structure.


```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
# Ver ausentes nas colunas críticas
cols_criticas = ['Employment', 'JobSat', 'RemoteWork']
print(df[cols_criticas].isnull().sum())

# Imputar com moda (valor mais frequente)
for col in cols_criticas:
    moda = df[col].mode()[0]
    df[col] = df[col].fillna(moda)

print("\nApós imputação:")
print(df[cols_criticas].isnull().sum())
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
# Converter YearsCodePro para numérico
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Criar faixas de experiência
bins = [0, 5, 10, 20, 100]
labels = ['0-5 anos', '5-10 anos', '10-20 anos', '+20 anos']
df['ExperienceRange'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels)

# Converter JobSat para numérico
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Calcular mediana de satisfação por faixa
satisfacao = df.groupby('ExperienceRange', observed=True)['JobSat'].median()
print(satisfacao)

# Visualizar
plt.figure(figsize=(8, 5))
satisfacao.plot(kind='bar', color='steelblue')
plt.title('Satisfação mediana por faixa de experiência')
plt.xlabel('Faixa de experiência')
plt.ylabel('Satisfação mediana')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

#### Step 5: Visualize Job Satisfaction


Use a count plot to show the distribution of `JobSat` values. This provides insights into the overall satisfaction levels of respondents.


```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='JobSat', order=df['JobSat'].value_counts().index, palette='Blues_d')
plt.title('Distribuição de satisfação no trabalho')
plt.xlabel('Nível de satisfação')
plt.ylabel('Contagem')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
# Count plot de RemoteWork
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='RemoteWork', palette='Set2')
plt.title('Distribuição de modalidade de trabalho')
plt.xlabel('Modalidade')
plt.ylabel('Contagem')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Cross-tabulation RemoteWork x Employment
df['Employment_Simple'] = df['Employment'].str.split(';').str[0]
crosstab = pd.crosstab(df['Employment_Simple'], df['RemoteWork'])
print(crosstab)
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
# Filtrar top 5 países
top_paises = df['Country'].value_counts().head(5).index
df_top = df[df['Country'].isin(top_paises)]

# Separar linguagens
languages = df_top['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
lang_country = df_top[['Country']].join(languages.rename('Language'))
lang_count = lang_country.groupby(['Country', 'Language']).size().reset_index(name='Count')

# Top 10 linguagens
top_langs = lang_count.groupby('Language')['Count'].sum().nlargest(10).index
lang_filtered = lang_count[lang_count['Language'].isin(top_langs)]

# Heatmap
pivot = lang_filtered.pivot(index='Language', columns='Country', values='Count').fillna(0)
plt.figure(figsize=(12, 7))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Top 10 linguagens por país')
plt.tight_layout()
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
df_scatter = df[['YearsCodePro', 'JobSat']].dropna()

plt.figure(figsize=(8, 5))
plt.scatter(df_scatter['YearsCodePro'], df_scatter['JobSat'], alpha=0.3, color='steelblue')
plt.title('Experiência vs Satisfação no trabalho')
plt.xlabel('Anos de experiência profissional')
plt.ylabel('Satisfação')
plt.tight_layout()
plt.show()

print("Correlação:", df_scatter.corr().loc['YearsCodePro', 'JobSat'].round(3))
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
df['EdLevel_Simple'] = df['EdLevel'].str.slice(0, 30)
df['Employment_Simple'] = df['Employment'].str.split(';').str[0]

crosstab_ed = pd.crosstab(df['EdLevel_Simple'], df['Employment_Simple'])
print(crosstab_ed)

plt.figure(figsize=(14, 7))
sns.heatmap(crosstab_ed, annot=True, fmt='d', cmap='Blues')
plt.title('Nível educacional vs Tipo de emprego')
plt.tight_layout()
plt.show()
```
